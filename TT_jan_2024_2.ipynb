```python
# function
def dfs_tabs(df_list, sheet_list, file_name):
    writer = pd.ExcelWriter(file_name,engine='xlsxwriter')   
    for dataframe, sheet in zip(df_list, sheet_list):
        dataframe.to_excel(writer, sheet_name=sheet, startrow=0 , startcol=0)   
    writer.save()
```

```python
Unnamed_cols = ["Unnamed: 1", "Unnamed: 2", 	"Unnamed: 3", 	"Unnamed: 4",
           "Unnamed: 5", "Unnamed: 6", 	"Unnamed: 7", 	"Unnamed: 8",
           "Unnamed: 9", 	"Unnamed: 10", 	"Unnamed: 11", 	"Unnamed: 12"]
# df["Unnamed: 1"][0]
```

```python
def dfs_for_csv(df):
    ## Remove 'NAN' from df
    df2=df.fillna(method="ffill")
    df2.dropna()
    ## day and times
    df_day_and_times = df2.iloc[0].values
    day_and_times = [df_day_and_times]+twenty4hour_times
    day_data = []
    for unnamed_col in Unnamed_cols:
        df_time = df2[unnamed_col][0]
        time_index = list(df_day_and_times).index(df_time)
        time = day_and_times[time_index]
        df_day = df2.iloc[0].values[0]
        for j in range(1,len(ROOMS)):
            room=df2.iloc[j].values[0]
            if df2[unnamed_col][j]!=df_time:
                if df2[unnamed_col][j]!='-':
                    code=df2[unnamed_col][j]
                    if code in  UNIVERSAL:
                        code_idx = UNIVERSAL.index(code)
                        lecturer = UNIVERSAL_LECTS[code_idx]
                    else:
                        lecturer = 'To check'
                    day_data.append({'campus':11, 
                                     'programme':'',
                                     'Stage':'',
                                     'unit':code,
                                     'day': df_day, 
                                     'period': time,
                                     '':'',
                                     'room': room, 
                                     'lecturer':lecturer})
        for k in range(len(LABS)+1):
            room=df2.iloc[j+k+1].values[0]
            if df2[unnamed_col][j+k+1]!=df_time:
                if df2[unnamed_col][j+k+1]!='-':
                    code=df2[unnamed_col][j+k+1]
                    if code in  UNIVERSAL:
                        code_idx = UNIVERSAL.index(code)
                        lecturer = UNIVERSAL_LECTS[code_idx]
                    else:
                        lecturer = 'To check'
                    day_data.append({'campus':11, 
                                     'programme':'',
                                     'Stage':'',
                                     'unit':code,
                                     'day': df_day, 
                                     'period': time,
                                     '':'',
                                     'room': room, 
                                     'lecturer':lecturer})
                    
    df_csv_sources = [pd.DataFrame([day_data[i]]) 
                     for i in range(len(day_data))]
    
    return pd.concat(df_csv_sources)
```

```python
def make_exam_tt(codes=UNIVERSAL):
    CODES = []
    for code in codes:
        if code not in CODES:
            CODES.append(code)
            
    SCHEDULES = []
    VENUES = ROOMS+LABS[:-2]
    
    for code in CODES:
        if code not in UNIVERSAL:
            print(code)
            lecturer = 'To check'
        else:
            code_idx = UNIVERSAL.index(code)
            lecturer = UNIVERSAL_LECTS[code_idx]
            title = UNIVERSAL_TITLES[code_idx]
            room = random.choice(UNIVERSAL_ROOMS[code_idx])
        SCHEDULES.append({'VENUE':room,
                          'CODE':code,
                          'TITLE':title, 
                          'SCHOOL':'SSAS', 
                          'No.s':'-',
                          'CHIEF INVIG':lecturer
                         })
        exam_dfs = [pd.DataFrame([SCHEDULES[i]]) 
                 for i in range(len(SCHEDULES))]
    
    return pd.concat(exam_dfs)
```
